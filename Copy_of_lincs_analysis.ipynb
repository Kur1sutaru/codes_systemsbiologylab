```python
data_lincs = pd.read_csv('lincsgenesearch (2).csv')
```

```python
data_lincs.head()
```

```python
data_lincs.shape
```

```python
cod_drugs= data_lincs['Compounds'].value_counts().index[0:10]
```

```python
cod_drugs
```

```python
data_lincs.query('Compounds in @cod_drugs')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(35, 12))
ax = sns.countplot(x = 'Compounds', data=data_lincs.query('Compounds in @cod_drugs'), order= data_lincs['Compounds'].value_counts().index[0:10], palette='husl')
#ax.set_title('Top compounds in MMD signature')
plt.title("Top compounds in MMD signature", fontsize = 20)
#plt.xlabel("genes", fontsize = 18)
plt.ylabel("genes", fontsize = 18)
plt.show()
```

```python
cod_cells = data_lincs['cell_line'].value_counts().index[0:10]
```

```python
cod_cells
```

```python
data_lincs.query('cell_line in @cod_cells')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(35, 12))
ax = sns.countplot(x = 'cell_line', data=data_lincs.query('cell_line in @cod_cells'), order= data_lincs['cell_line'].value_counts().index[0:10], palette='husl')
ax.set_title('Top cell lines')
plt.title("Top cell lines", fontsize = 20)
#plt.xlabel("Top drugs", fontsize = 18)
plt.ylabel("genes", fontsize = 18)
plt.show()
```
