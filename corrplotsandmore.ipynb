```python
data = pd.read_csv('multiplex_RNASEH2B.csv')
```

```python
data.shape
```

DataFrame.corr(method='pearson', min_periods=1)[DEFAULT] Compute pairwise correlation of columns, excluding NA/null values.

Parameters method{‘pearson’, ‘kendall’, ‘spearman’} or callable Method of correlation:

pearson : standard correlation coefficient

kendall : Kendall Tau correlation coefficient

spearman : Spearman rank correlation

```python
# default = pearson
correlation = data.corr()
```

```python
correlation
```

```python
# correlation matrix plot
import seaborn as sns
plot = sns.heatmap(correlation, annot = True, fmt=".1f", linewidths=.8)
plot
```

```python
# testing the kendall correlation
correlationkendall = data.corr(method= "kendall")
```

```python
correlationkendall
```

```python
import seaborn as sns
plot = sns.heatmap(correlationkendall, annot = True, fmt=".1f", linewidths=.8)
plot
```

```python
# testing the spearman correlation
correlationspearman = data.corr(method= "spearman")
```

```python
correlationspearman
```

```python
# correlation matrix plot
import seaborn as sns
plot = sns.heatmap(correlationspearman, annot = True, fmt=".1f", linewidths=.8)
plot
```
